```julia
using Plots
```

```julia
x = range(-π, π; length = 20)
xlim = extrema(x) .* 1.1
```

```julia
y = @. sin(x) + 0.3randn()
X = x .^ (0:3)'
β̂ = X \ y
f(x) = evalpoly(x, β̂)
scatter(x, y; label = "data", color = 1, legend = :topleft)
plot!(sin, xlim...; label = "y = sin(x)", color = 1)
plot!(f, xlim...; label = "degree-3 polynomial", color = 2, lw = 2)
```
