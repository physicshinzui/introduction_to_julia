In this notebook, we handle the attribute `seriestype` by which we can change a plotting style, e.g., line plot to scatter plot.


```julia
using Plots
```

```julia
gr()
x = 1:10; y = rand(10,2)
p = plot(x,y, seriestype=:scatter, title="Scatter plot")
xlabel!(p, "Steps")
ylabel!(p, "Value")
```
