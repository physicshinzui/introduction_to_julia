## Backend Ex 1. Plotly
Firstly, we shall set `Plotly` as the backend we'll use.

```julia
x = 1:10; y = rand(10,2)
plotly() # Set the backend to Plotly
plot(x,y, title="This is plotted by Plotly.")
```

## Backend Ex 2. GR 

```julia
gr()
plot(x,y, title="This is plotted by GR.")
```
