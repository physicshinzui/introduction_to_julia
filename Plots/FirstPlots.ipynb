This notebok introduces the most basic plot: line plotting.

Note: 
This notebook is a jupyter-based one, so any plot is displayed automatically, but if we want to display plots in a script, then we must add the commmand`display(plot(x, y))`.

```julia
# import Plot
using Plots
```

## Line plot: 1D random walk

```julia
r = rand(10)
x = 1:10
plot(x, r, label="1D random walk", aspect_ratio=1)
```

Look at how 1:10 works here. 

```julia
typeof(x)
```

This says that `x` is in the type of `UnitRange{Int64}`.

## Multiple lines: Implicit way

```julia
r1 = rand(10)
r2 = rand(10)
x  = 1:10 
plot(x, r1)
plot!(x, r2) # "!" means that a new plot is added to the current plot: The global `Plots.CURRENT_PLOT`. 
```

## Multiple Lines: Explicit way

```julia
p = plot(x, r1) # Declare a variable of `plot`
plot!(p, r2)
```

## Multiple Lines from a matrix

```julia
rxy = rand(10,2) # Suppose that each column is a sequence we want to plot individually.
```

```julia
x = 1:10
plot(x, rxy) # plot([x range], matrix[10,2])
# This plots each column sequence in a shared x axis.
```

## Multiple Plots with some modifications

```julia
r = rand(10,3)
x = 1:10
plot(x, r, title="Random Walk", lw=3, labels=["Line1" "Line2" "Line3"]) 
# Note that there is no comma among label's names, which I often make a mistake for.
# lw: stands for linewidth, by which we can control line width.
```

If we want to add a x and y labels onto the figure, then try the following (this is the implicit way):

```julia
xlabel!("Steps")
ylabel!("Value")
```

If you want to do this by specifying which plot, you can do like:

```julia
r = rand(10,3)
x = 1:10
# --- Explicitly declare a plot variable, and then is put to xlabel and ylabel functions. 
p = plot(x, r, title="Random Walk", lw=3, labels=["Line1" "Line2" "Line3"]) 
xlabel!(p, "Steps")
ylabel!(p, "Steps")
```
