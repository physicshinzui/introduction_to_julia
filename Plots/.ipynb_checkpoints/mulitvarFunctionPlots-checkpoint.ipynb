```julia
using Plots
#plotly()
gr()
```

```julia
x, y = -10:10, -10:10
```

```julia
f(x,y) = 2x^2 + 12x*y + 7y^2
```

```julia
z = [f(xi, yi) for xi in x, yi in y]
```

```julia
p1 = plot(x,y,z,st=:wireframe)
p2 = plot(x,y,z,st=:contour)
p = plot(p1, p2, layout=(1,2) )
```
