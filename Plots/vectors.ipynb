## 1. Draw two vectors from the origin.
Usage: 
    `plot([begin coordinates], [end coordinates])`

```julia
# Define two unit vectors
e1 = [1,0]; e2 = [0,1]

e1_begin = [[0,e1[1]]]; e1_end = [[0,e1[2]]]
e2_begin = [[0,e2[1]]]; e2_end = [[0,e2[2]]]

plot(e1_begin, e1_end, label="e1", arrow=true, lw=3, color=:black, aspect_ratio=1)
plot!(e2_begin, e2_end, label="e2", arrow=true, lw=3, color=:black)
```

### Another example: Three vectors

```julia
vecs = [1 2 3; 4 5 6; 7 8 9]
x1 = vecs[:,1]
x2 = vecs[:,2]
x3 = vecs[:,3]

plot([0,x1[1]],[0,x1[2]], arrow=true)
plot!([0,x2[1]],[0,x2[2]], arrow=true)
plot!([0,x3[1]],[0,x3[2]], arrow=true)
```

## 2. 2D Vector field
Usage: 
`quiver(x, y, quiver=(u,v))`
- x : Start points of x coordinate of vectors [x1, x2,...,xn]
- y : Start points of y coordinate of vectors [y1, y2, ..., yn]
- u : x-axis components of vectors [u1, u2, ..., un]. 
- v : y-axis components of vectors [v1, v2, ..., vn]. 

Each vector (ui, vi) is depicted from (xi,yi). For example, 

```julia
quiver([1,2],[3,2],quiver=([1,1],[1,2]), lw=3, color=:black)
plot!([1.0,2.0],[3.0,2.0], seriestype=:scatter, label=nothing, color=:black) # for starting points
annotate!(1.5,3.0, text("From (1,3),\n the vector (1,1) was depicted."))
```
